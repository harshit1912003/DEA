### Time Complexity Analysis wrt n (Number of DMUs)

```python
times_n = []
ns = []
max_n = 100
for n in range(1, max_n + 1, 2):
    x, y = initializeZ(n, 5, 5)
    dea = DEA(x, y)

    start_time = time.time()
    dea.fdh_input_vrs()
    end_time = time.time()

    times_n.append(end_time - start_time)
    ns.append(n)
```

```python
plt.plot(ns, times_n, marker='o')

plt.xlabel('n (Data size)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of DEA.fdh_input_vrs()')
plt.grid(True)
plt.show()
```
