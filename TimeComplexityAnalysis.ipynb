### Importing dependencies and necessary packages

```python
from models.modelsDEA import DEA
from models.modelsFDH import FDH
import numpy as np
import time
import matplotlib.pyplot as plt
from utils.datainput import initializeUnif
```

# DEA

### Time Complexity Analysis wrt n (Number of DMUs)

```python
times_n = []
ns = []
max_n = 100
for n in range(1, max_n + 1, 2):
    x, y = initializeUnif(n, 5, 5, 1, 21)
    dea = DEA(x, y)
    start_time = time.time()
    dea.bcc_input()
    end_time = time.time()
    times_n.append(end_time - start_time)
    ns.append(n)
```

```python
plt.plot(ns, times_n, marker='o')

plt.xlabel('n (Number of DMUs)')
plt.ylabel('Time (seconds)')
plt.title('Time Complexity of DEA.bcc_input()')
plt.grid(True)
plt.show()
```
