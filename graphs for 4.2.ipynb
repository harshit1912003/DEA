```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.uniform(1, 10, 50) 
fx = np.log(x) + 3
u = np.abs(np.random.normal(0, 0.4, 50))

fx_obs = fx - u

plt.figure(figsize=(10, 5))
plt.scatter(x, fx_obs, color='blue', label="Scenario A (Observed)")
plt.plot(np.sort(x), np.sort(fx), color='blue', linestyle="dashed", linewidth=2, label="True Frontier A")

plt.xlabel("Input (x)")
plt.ylabel("Output (y)")
plt.title("Frontier Functions for Scenarios A and B")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Set random seed for reproducibility
np.random.seed(42)

# Create a meshgrid for x1 and x2 in the range [1, 10]
x1 = np.linspace(1, 10, 50)
x2 = np.linspace(1, 10, 50)
X1, X2 = np.meshgrid(x1, x2)

# Deterministic production frontier for Scenario 2:
# f(x1, x2) = 0.1*x1 + 0.1*x2 + 0.3*sqrt(x1*x2)
F = 0.1 * X1 + 0.1 * X2 + 0.3 * np.sqrt(X1 * X2)

# Simulate observed data points
n = 100  # sample size for illustration
sample_x1 = np.random.uniform(1, 10, n)
sample_x2 = np.random.uniform(1, 10, n)
sample_f = 0.1 * sample_x1 + 0.1 * sample_x2 + 0.3 * np.sqrt(sample_x1 * sample_x2)

# Inefficiency term: absolute value of N(0, 0.4)
u = np.abs(np.random.normal(0, 0.4, n))
sample_y = sample_f - u

# Plotting the production frontier and the observed data points
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the deterministic frontier surface
surf = ax.plot_surface(X1, X2, F, cmap='viridis', alpha=0.7, edgecolor='none')
cbar = fig.colorbar(surf, shrink=0.5, aspect=5, label='Frontier Output')

# Overlay scatter plot for the observed data points (with noise)
ax.scatter(sample_x1, sample_x2, sample_y, color='red', label='Observed Data', s=50)

ax.set_xlabel('Input $x_1$')
ax.set_ylabel('Input $x_2$')
ax.set_zlabel('Output $y$')
ax.set_title('3D Plot of Production Frontier for Scenario 2')
ax.legend()

plt.show()
```
