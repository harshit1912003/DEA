```python
from gurobipy import Model, GRB, quicksum
import pandas as pd

def nonhomofdh(n, T, m, s, a, d, x, y):
    results = []

    base_model = Model("FDH_Input_CRS")
    base_model.setParam(GRB.Param.OutputFlag, 0)

    for k in range(n):  
        model = base_model.copy()

        lambda_t = model.addVars(T, vtype=GRB.BINARY, name="lambda")
        phi = model.addVar(lb=0, name="phi")  

        model.setObjective(phi, GRB.MAXIMIZE)

        for j in range(m):
            model.addConstr(
                quicksum(lambda_t[t] * a[t, j] for t in range(T)) <= x[k, j]
            )

        for r in range(s):
            model.addConstr(
                quicksum(lambda_t[t] * d[t, r] for t in range(T)) >= phi * y[k, r]
            )

        model.addConstr(quicksum(lambda_t[t] for t in range(T)) == 1)

        model.optimize()

        if model.Status == GRB.OPTIMAL:
            results.append({
                "DMU": k + 1,  # DMU index (1-based)
                "efficiency": model.objVal
            })
        else:
            results.append({
                "DMU": k + 1,  # DMU index (1-based)
                "efficiency": None  # Record that no solution was found
            })

            
    results = pd.DataFrame(results)
    return results
```
